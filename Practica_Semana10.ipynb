```python
import pandas as pd
from google.colab import files

# La Unidad de Control brinda los datos del Tllegada y Tcpu desde la RAM

Tllegada = [6, 3, 4, 1, 2, 5]  # Cola de instrucciones
Tcpu = [1, 3, 6, 5, 2, 1]  # Tiempos de cpu usados por cada instrucción
"""
# Ejemplo 2:
Tllegada = [3, 5, 7, 1, 4, 2, 6]  
Tcpu = [2, 3, 8, 2, 4, 1, 1]  
"""

Tllegada2 = list(Tllegada) # array temporal para ejecutar el orden de instrucciones
pos = [] # aquí se ordena de forma ascendente la ejecución de instrucciones
for i in range(0, len(Tllegada)):
    pos.append(0)

# Ordena los procesos a ejecutar con respecto a la cola de instrucciones
i = 0
for lista in range(0, len(Tllegada2)):
    minimo = min(Tllegada2)
    i_min = Tllegada2.index(minimo)
    pos[i] = i_min
    Tllegada2[i_min] = 1000
    i = i + 1
print(pos)  # Imprime el array de los procesos
```

```python
datos = []  # Se inicia la lista de diccionarios para luego convertirlo a dataframe

# Información del ciclo de reloj
ciclo_reloj = 0
interaccion = "Init"
process_running = "none"
tcpu_running = "none"
process_submit = pos[0]
tcpu_submit = Tcpu[process_submit]

# Función para agregar un diccionario a la lista de diccionarios
def agregarfila(): 
    # Se genera un diccionario
    fila = {
        "Clock": ciclo_reloj,
        "Interaction": interaccion,
        "Process running": process_running,
        "T CPU - P. running": tcpu_running,
        "Submit process": process_submit,
        "# T CPU - Submit": tcpu_submit,
    }

    datos.append(fila)  # Se añade el diccionario a la lista de diccionarios datos

agregarfila() # Se agrega diccionario

# cache L1 y 1 core de CPU

# Recorre el array de instrucciones, pertenece a la columna proceso
for clock in range(0, len(Tllegada) + 1):
    # Recorre el array de los ciclos de reloj con que ejecuta cada instrucción, pertenece a la columna tcpu
    for i in range(0, len(Tcpu)):
        po = pos[i]
        try:
            # (FETCH, reconoce por cada instrucción cuantos ciclos de reloj hay)
            if clock == Tllegada[po]:
                for j in range(0, Tcpu[po]):
                    # Se actualiza la información del ciclo de reloj
                    ciclo_reloj += 1
                    interaccion = i
                    process_running = po
                    tcpu_running = j

                    try:
                        process_submit = pos[i + 1]
                        tcpu_submit = Tcpu[process_submit]

                        agregarfila() # Se agrega diccionario
                    except Exception as error:
                        # Ya no hay procesos submit
                        process_submit = "none"
                        tcpu_submit = "none"

                        agregarfila() # Se agrega diccionario
        except Exception as error:
            print("Caught this error: " + repr(error))

df = pd.DataFrame(datos) # La lista de diccionarios la transformamos a un dataframe 
df
```
